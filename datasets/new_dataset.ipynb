```python
#read dataset_15000_generated.csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Read the dataset
df = pd.read_csv('./dataset15000_generated.csv')
```

```python
#pick in the dataframe only the rows that contain "True,True,True,True" in the columns "req_1", "req_2", "req_3", "req_4"
all_true = df[(df['req_0'] == True) & (df['req_1'] == True) & (df['req_2'] == True) & (df['req_3'] == True)]
all_false = df[(df['req_0'] == False) & (df['req_1'] == False) & (df['req_2'] == False) & (df['req_3'] == False)]
#cut the all_false dataframe to have the same number of rows as all_true
all_false = all_false.sample(n=len(all_true), random_state=42)
```

```python
#create a new dataframe with all_true and all_false shuffled
new_df = pd.concat([all_true, all_false]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
new_df.head()
```

```python
#export  it in csv file
new_df.to_csv('./new_dataset.csv', index=False)
```

```python
#pick in the dataframe only the rows that contain "True,True,True,True" in the columns "req_1", "req_2", "req_3", "req_4"
one_true = df[(df['req_1'] == True)]
one_false = df[(df['req_1'] == False)]
```

```python
#cut the all_false dataframe to have the same number of rows as all_true
one_false = one_false.sample(n=len(one_true), random_state=42)
#remove from all_true the rows that are in one_true
all_true = all_true[~all_true.index.isin(one_true.index)]
#create a new dataframe with all_true and all_false shuffled
new_df = pd.concat([one_true, one_false, all_true]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
new_df.head()
```

```python
#export  it in csv file
new_df.to_csv('./new_dataset.csv', index=False)
```

```python
#count true and false in each column
print(new_df['req_0'].value_counts())
print(new_df['req_1'].value_counts())
print(new_df['req_2'].value_counts())
print(new_df['req_3'].value_counts())

#count the number of rows for which all the req_0, req_1, req_2, req_3 are False
print(len(new_df[(new_df['req_0'] == False) & (new_df['req_1'] == False) & (new_df['req_2'] == False) & (new_df['req_3'] == False)]))
```
